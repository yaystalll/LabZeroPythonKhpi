```python
import math

def is_sum_of_squares(N):
    for a in range(0, int(math.isqrt(N)) + 1):
        b = math.isqrt(N - a**2)
        if a**2 + b**2 == N:
            return a, b
    return None

def task_10():
    try:
        N = int(input("Введіть число N: "))
        result = is_sum_of_squares(N)
        
        if result:
            a, b = result
            print(f"{a}^2 + {b}^2 = {N}")
        else:
            print(f"Число {N} не можна представити квадратами цілих чисел.")
    
    except ValueError:
        print("Помилка! Будь ласка, введіть ціле число.")
```

```python
if __name__ == "__main__":
    task_10()
```
